```python
from datetime import datetime

def parse_date(date_str):
    """Функция для парсинга строки с датой в объект datetime."""
    formats = [
        "%A, %B %d, %Y",      # Wednesday, October 2, 2002
        "%A, %d.%m.%y",       # Friday, 11.10.13
        "%A, %d %B %Y"        # Thursday, 18 August 1977
    ]

    for fmt in formats:
        try:
            return datetime.strptime(date_str, fmt)
        except ValueError:
            continue
    raise ValueError(f"Не удалось распознать формат даты: {date_str}")

print("Введите даты для преобразования (формат: 'Wednesday, October 2, 2002' и др.)")
print("Для завершения введите 'quit'.")

while True:
    user_input = input("Введите дату: ")

    if user_input.lower() == 'quit':
        print("Программа завершена.")
        break

    try:
        dt = parse_date(user_input)
        print(f"Результат: {dt}")
    except ValueError as e:
        print(f"Ошибка: {e}")
```
