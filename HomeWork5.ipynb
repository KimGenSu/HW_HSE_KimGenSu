```python
from datetime import datetime

date_str = "Wednesday, October 2, 2002"
date = datetime.strptime(date_str, "%A, %B %d, %Y")
print("The Moscow Times -",date)

date_str = "Friday, 11.10.13"
date = datetime.strptime(date_str, "%A, %d.%m.%y")
print("The Guardian — Friday -", date)

date_str = "Thursday, 18 August 1977"
date = datetime.strptime(date_str, "%A, %d %B %Y")
print("Daily News - ", date)
```
