```python
import csv

class Pokupatel:

    def __init__(self, fio, ustroistvo, brauzer, pol, vozrast, summa, region):
        self.fio = fio
        self.ustroistvo = ustroistvo
        self.brauzer = brauzer
        self.pol = pol
        self.vozrast = int(vozrast)
        self.summa = int(summa)
        self.region = region

    def opredelit_pol(self):
        if self.pol == 'female':
            return 'женского'
        else:
            return 'мужского'

    def kak_nazvat_ustroistvo(self):
        slovar = {
            'mobile': 'мобильного',
            'tablet': 'планшета',
            'laptop': 'ноутбука',
            'desktop': 'компьютера'
        }
        if self.ustroistvo in slovar:
            return slovar[self.ustroistvo]
        return self.ustroistvo

    def sdelat_opisanie(self):
        pol_tekst = self.opredelit_pol()
        ustroistvo_tekst = self.kak_nazvat_ustroistvo()

        if self.pol == 'female':
            glagol = 'совершила'
        else:
            glagol = 'совершил'

        if self.region == '-':
            region_tekst = 'не указан'
        else:
            region_tekst = self.region

        opisanie = (f"Пользователь {self.fio} {pol_tekst} пола, "
                   f"{self.vozrast} лет {glagol} покупку на {self.summa} у.е. "
                   f"с {ustroistvo_tekst} браузера {self.brauzer}. "
                   f"Регион, из которого совершалась покупка: {region_tekst}.")

        return opisanie


def prochitat_fail(imya_faila):
    spisok_pokupateley = []

    try:
        with open(imya_faila, 'r', encoding='utf-8') as f:
            chitatel = csv.DictReader(f)

            nomer_stroki = 0
            for stroka in chitatel:
                nomer_stroki += 1

                try:
                    chelovek = Pokupatel(
                        fio=stroka['name'].strip(),
                        ustroistvo=stroka['device_type'].strip(),
                        brauzer=stroka['browser'].strip(),
                        pol=stroka['sex'].strip(),
                        vozrast=stroka['age'].strip(),
                        summa=stroka['bill'].strip(),
                        region=stroka['region'].strip()
                    )
                    spisok_pokupateley.append(chelovek)
                except:
                    # Если какая-то строка нечитаемая - пропускаем
                    continue

            print(f"Прочитано строк из файла: {nomer_stroki}")
            print(f"Успешно обработано: {len(spisok_pokupateley)}")

    except FileNotFoundError:
        print(f"Файл {imya_faila} не найден!")
        return []

    return spisok_pokupateley


def sohranit_v_fail(spisok, imya_vihodnogo_faila):
    if not spisok:
        print("Нет данных для сохранения")
        return False

    try:
        with open(imya_vihodnogo_faila, 'w', encoding='utf-8') as f:
            vsego_zapisano = 0

            for i, pokupatel in enumerate(spisok):
                tekst = pokupatel.sdelat_opisanie()
                f.write(tekst)
                f.write('\n\n')

                # Разделитель между записями, кроме последней
                if i < len(spisok) - 1:
                    f.write('-' * 70 + '\n\n')

                vsego_zapisano += 1

            print(f"Сохранено записей в файл: {vsego_zapisano}")
            return True

    except Exception as e:
        print(f"Ошибка при записи: {e}")
        return False


def pokazat_primeri(spisok, skolko=2):
    if not spisok:
        print("Нет данных для показа")
        return

    print("\nПримеры готовых описаний:")
    print("="*60)

    for i in range(min(skolko, len(spisok))):
        print(f"\nПример {i+1}:")
        print(spisok[i].sdelat_opisanie())
        print("-"*60)


def proverit_fail(imya_faila):
    try:
        with open(imya_faila, 'r', encoding='utf-8') as f:
            soderjimoe = f.read(500)  # Первые 500 символов
            if soderjimoe:
                print(f"\nНачало файла {imya_faila}:")
                print("-"*40)
                print(soderjimoe[:300] + "...")
                return True
    except:
        pass

    return False


def glavnaya_funkciya():
    print("Обработка данных о покупателях")
    print("="*50)

    # Имя файла с данными
    csv_fail = "web_clients_correct.csv"

    # Шаг 1 - чтение файла
    print("\n1. Читаю данные из CSV...")
    vse_pokupateli = prochitat_fail(csv_fail)

    if not vse_pokupateli:
        print("Не получилось прочитать файл")
        return

    # Шаг 2 - показать примеры
    print("\n2. Формирую описания...")
    pokazat_primeri(vse_pokupateli)

    # Шаг 3 - сохранить в файл
    print("\n3. Сохраняю в текстовый файл...")
    txt_fail = "opisaniya_pokupateley.txt"
    uspeh = sohranit_v_fail(vse_pokupateli, txt_fail)

    if uspeh:
        print(f"\nГотово! Файл сохранён: {txt_fail}")

        # Проверим, что файл создан
        proverit_fail(txt_fail)
    else:
        print("Не удалось сохранить файл")


# Точка входа
if __name__ == "__main__":
    glavnaya_funkciya()
```
